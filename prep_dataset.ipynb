# Weather data

```python
data = pd.read_csv('Datasett/veardata/Veardata_Bjornfjell.csv', sep=';')
```

```python
data
```

```python
data.MVT_ID.unique()
```

```python
data['VALUE'] = data['VALUE'].map(lambda x: float(x.replace(',', '.')))
```

```python
# extract data for 'UNIT'
def convert_to_column(unitstr, data):
    unit = data[data['DESCRIPTION'] == unitstr].copy()
    unit.rename(columns={'VALUE': unitstr}, inplace=True)
    return unit[[unitstr, 'MEASURE_TIME']]
```

```python
unit_df = []
for unitstr in data.DESCRIPTION.unique():
    unit_df.append(convert_to_column(unitstr, data))
```

```python
data
```

```python
data.dtypes
```

```python
data_time = data[['MEASURE_TIME']].copy()
```

```python
for unit in unit_df:
    data_time = pd.merge(left=data_time, right=unit, left_on='MEASURE_TIME', how='left', right_on='MEASURE_TIME')
```

```python
data_time
```

```python
data_time.dtypes
```

```python
data.to_csv('./data/vaerdata_bjornfjellet.csv', index=False)
```

# AKT målepunkter from Vegkart

```python
import geopandas as gpd
import nvdbapiv3 as nvdb
from nvdbapiv3 import nvdb2geojson
```

```python
# We need all on the EV6
q = nvdb.nvdbFagdata(482) # Trafikkregistreringsstasjon
q.filter({'egenskap': '(5201=7081 OR 5201=12987) AND (9293=12992) AND (3910=4892)'})
```

```python
q_geo = nvdb2geojson.fagdata2geojson(q)
qdf = gpd.GeoDataFrame.from_features(q_geo['features'])
qdf.set_crs(epsg=25833)
```

```python
qdf = qdf.drop(columns=['Geometri, punkt', 'Vedlikeholdsansvarlig', 'Eier', 'Prosjektreferanse', 'kryssystem', 'Sist oppdatert fra MSR', 'Tilleggsinformasjon', 'sideanlegg', 'Sist oppdatert fra NorTraf',
                       'Mot-retning (fra)', 'Med-retning (fra)', 'Trafikantgruppe', 'antall vegsegmenter' , 'retning', 'typeVeg', 'typeVeg_sosi',
                       'vegsegment nr'])
```

```python
qdf = qdf[qdf['kortform'].str.contains("EV6")]
```

```python
qdf = qdf.set_crs(epsg=25833, allow_override=True)
```

```python
qdf.crs
```

```python
qdf.to_file('./data/trafikkregistreringsstasjoner_e6.geojson', driver='GeoJSON')
```
