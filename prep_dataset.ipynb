```python
import pandas as pd
```

```python
data = pd.read_csv('Datasett/Trafikkregistreringsdatasett/trafikkregistreringspunkt 1 - uke 42 og 43 2020.csv', sep=';')
```

```python
data['length'] = data['length'].map(lambda x: float(x.replace(',', '.')))
```

```python
data['qspeed'] = data['qspeed'].map(lambda x: float(x.replace(',', '.')))
```

```python
data['speed'] = data['speed'].map(lambda x: float(x.replace(',', '.')))
```

```python
data['time_gap'] = data['time_gap'].map(lambda x: float(x.replace(',', '.')))
```

```python
data['_score'].unique()
```

```python
data.drop(columns=['_score'], inplace=True)
```

```python
data['_type'].unique()
```

```python
data.drop(columns=['_type'], inplace=True)
```

```python
data['county_id'].unique()
```

```python
data.drop(columns=['county_id'], inplace=True)
```

```python
data['datalogger_type'].unique()
```

```python
data.drop(columns=['datalogger_type'], inplace=True)
```

```python
data['firmware_version'].unique()
```

```python
data['_index'].unique()
```
